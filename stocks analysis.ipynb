```python
import numpy as np
import pandas as pd
from pandas_datareader import data
import altair as alt
```

```python
start = '2019-1-1'
end = '2020-12-31'
source = 'yahoo'
apple = data.DataReader("AAPL", start=start ,end=end, data_source=source).reset_index()
ibm = data.DataReader("IBM", start=start ,end=end, data_source=source).reset_index()
microsoft = data.DataReader("MSFT", start=start ,end=end, data_source=source).reset_index()
```

```python
apple.head()
```

```python
apple['Symbol'] = 'AAPL'
ibm['Symbol'] = 'IBM'
microsoft['Symbol'] = 'MSFT'

stocks = pd.concat([apple[['Date','Close','Volume','Symbol']],ibm[['Date','Close','Volume','Symbol']],
     microsoft[['Date','Close','Volume','Symbol']]],axis=0)

stocks.head()
```

```python
# ALTAIR uses shown below, stock data visualization for analysis
```

```python
(alt.Chart(stocks[stocks.Symbol == 'AAPL'],title='Apple Stock Closing Price by Date').mark_line().encode(x='Date', y='Close'))
```

```python
(alt.Chart(stocks,title='Apple, Microsoft & IBM Stock Closing Prices By Date').mark_line().encode(x='Date',y='Close',color='Symbol').properties(height=400,width=600))
```

```python
# In below cell -> 7D is for 7(seven) days
```

```python
msft_resampled = stocks[stocks.Symbol == 'MSFT'].resample('7D', on='Date').mean().reset_index()
aapl_resampled = stocks[stocks.Symbol == 'AAPL'].resample('7D', on='Date').mean().reset_index()
ibm_resampled = stocks[stocks.Symbol == 'IBM'].resample('7D', on='Date').mean().reset_index()

aapl_resampled['Symbol'] = 'AAPL'
msft_resampled['Symbol'] = 'MSFT'
ibm_resampled['Symbol'] = 'IBM'
resampled_stocks = pd.concat([msft_resampled[['Date','Close','Volume','Symbol']],
                              aapl_resampled[['Date','Close','Volume','Symbol']],
                              ibm_resampled[['Date','Close','Volume','Symbol']]],axis=0)
```

```python
(alt.Chart(resampled_stocks,title='Use of RESAMPLE function on the stock prices').mark_line().encode(x='Date',y='Close',color='Symbol').properties(height=400,width=600))
```
